4. Write a Python program to calculate the factorial of a number provided by
the user using a while loop


```python

num = int(input("Enter a number: "))


factorial = 1
original_num = num

if num < 0:
    print("Factorial does not exist for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1.")
else:
    while num > 0:
        factorial *= num
        num -= 1

    print(f"The factorial of {original_num} is {factorial}")
```

8. Write Python code to create a class Person with attributes name and age,
and a method display that prints these attributes.


```python
class Person:
    def __init__(self, name, age):
        """
        Initializes the Person object with name and age.
        """
        self.name = name
        self.age = age

    def display(self):
        """
        Prints the attributes of the Person.
        """
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


person1 = Person("Arjun", 21)

person1.display()
```
